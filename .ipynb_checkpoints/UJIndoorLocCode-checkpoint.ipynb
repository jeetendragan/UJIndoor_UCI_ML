```python
training_data = pd.read_csv('trainingData.csv')
test_data = pd.read_csv('validationData.csv')
train_lat_long_X = training_data.drop(columns=["LONGITUDE", "LATITUDE", "FLOOR", "BUILDINGID", "SPACEID", "RELATIVEPOSITION", "USERID", "PHONEID", "TIMESTAMP"])
train_lat_long_Y = training_data.loc[:, "LONGITUDE":"LATITUDE"]
test_lat_long_X = test_data.drop(columns=["LONGITUDE", "LATITUDE", "FLOOR", "BUILDINGID", "SPACEID", "RELATIVEPOSITION", "USERID", "PHONEID", "TIMESTAMP"])
test_lat_long_Y = test_data.loc[:, "LONGITUDE":"LATITUDE"]
```

```python
training_data.columns
```

```python
def get_accuracy(predictions, actual_values):
  correct_classif = [1 if pred_i == actual_i else 0 for pred_i, actual_i in zip(predictions, actual_values)]
  correct_classif = np.array(correct_classif)
  return correct_classif.sum() / len(actual_values)
```

1.   Create an object of class Sequential(tf.keras.models)
2.   Add layers to it

```python
import tensorflow as tf
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(512, activation=tf.nn.relu),
  tf.keras.layers.Dense(2, activation = 'linear')
])
```

3.    Compile and specify the optimization algorithm, loss, and metrics.

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

4.   Fit/Train the model

```python
val_data = (test_lat_long_X, test_lat_long_Y)
```

```python
epochs = 200
result = model.fit(train_lat_long_X.values.astype(float), train_lat_long_Y.values.astype(float), validation_data = val_data, epochs=epochs)
```

5. Evaluate/Test the model

```python
import matplotlib.pyplot as plt
x_values = range(1,epochs+1)
plt.plot(x_values, result.history['loss'])
plt.title("Train Error")
plt.show()
```

```python
plt.plot(x_values, result.history['val_loss'])
plt.title("Test Error")
plt.show()
```

## With 724 nodes in the hidden layer

```python
model_2 = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(724, activation=tf.nn.relu),
  tf.keras.layers.Dense(2, activation = 'linear')
])
model_2.compile(optimizer='adam', loss='mean_squared_error')
epochs = 100
result_2 = model_2.fit(train_lat_long_X.values.astype(float), train_lat_long_Y.values.astype(float), validation_data = val_data, epochs=epochs)
```

```python
x_values = range(1,epochs+1)
plt.plot(x_values, result_2.history['loss'])
plt.title("Train Error")
plt.show()
```

```python
plt.plot(x_values, result_2.history['val_loss'])
plt.title("Test Error")
plt.show()
```

## Prediciting buildingid

```python
train_building_X = train_lat_long_X
train_building_Y = training_data.loc[:, "BUILDINGID"]
test_building_X = test_lat_long_X
test_building_Y = test_data.loc[:, "BUILDINGID"]
val_data_building = (test_building_X, test_building_Y)
```

```python
model_bldn_1 = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(724, activation=tf.nn.relu),
  tf.keras.layers.Dense(3, activation = 'softmax')
])
model_bldn_1.compile(optimizer='adam', 
                loss='sparse_categorical_crossentropy',
                metrics = ['accuracy'])
epochs = 100
result_2 = model_bldn_1.fit(train_building_X.values.astype(float), train_building_Y.values.astype(float), validation_data = val_data_building, epochs=epochs)
```

```python
x_values = range(1,epochs+1)
plt.title("Train Loss")
plt.plot(x_values, result_2.history['loss'])
plt.show()
```

```python
plt.title("Test Loss")
plt.plot(x_values, result_2.history['val_loss'])
plt.show()
```
